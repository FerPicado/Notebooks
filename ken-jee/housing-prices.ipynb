```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats
import plotly.graph_objects as go
import plotly.express as px
from IPython.display import HTML
```

```python
# dataframe
df = pd.read_csv("./data/train.csv")
df.head()
```

```python
# Function to create scrollable table within a small window
def create_scrollable_table(df, table_id, title):
    html = f'<h3>{title}</h3>'
    html += f'<div id="{table_id}" style="height:200px; overflow:auto;">'
    html += df.to_html()
    html += '</div>'
    return html
```

```python
df.shape
```

```python
# numerical values
numerical_features =  df.select_dtypes(include=[np.number])
numerical_features.head()
```

```python
numerical_features.describe()
```

```python
# Summary statistics for numerical features
summary_stats = numerical_features.describe().T
html_numerical = create_scrollable_table(df=summary_stats,
                                        table_id=numerical_features,
                                        title="Summary statistics for numerical features")
display(HTML(html_numerical))
```

```python
# summaary statistics for categorical features

cat_features = df.select_dtypes(include=[object], exclude=[np.number])
cat_summary_stats = cat_features.describe().T
html_cat = create_scrollable_table(df=cat_summary_stats,
                                  table_id="cat_features",
                                  title="Summary statistics for categorical features")
```

```python
display(HTML(html_cat))
```
